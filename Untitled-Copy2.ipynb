```python
qr0 = QuantumRegister(4, 'qr0')
rg0 = QuantumCircuit(qr0, name="Reversible Gate")
```

```python
rg0.reset(qr0[2])
rg0.reset(qr0[3])
rg0.cx(qr0[0],qr0[2])
rg0.cx(qr0[1],qr0[2])
rg0.x(qr0[1])
rg0.ccx(qr0[0],qr0[1],qr0[3])
rg0.x(qr0[1])
rg0.x(qr0[2])

rg0.draw(output='mpl')
```

```python
reg0 = rg0.to_instruction()
```
